### Wagen leeftijd

```python
# hoe ouder de wagen, hoe meer ongelukken met hindernissen buiten de rijbaan
# hoe ouder, hoe meer ongelukken met één bestuurder en zonder hindernissen
a = collision_df.groupby("TX_CLASS_VEHICLE_AGE_AT_COLL_YR").TX_COLLISON_TYPE_NL.value_counts(normalize=True).unstack()
a.plot.bar()
```

```python
# hoe ouder de wagen, hoe meer zwaar gewonden en doden
a = collision_df.groupby("TX_CLASS_VEHICLE_AGE_AT_COLL_YR").TX_ACCIDENT_CLASS_NL.value_counts(normalize=True).unstack()
a.plot.bar()
```

```python
a = collision_df.groupby("TX_CLASS_VEHICLE_AGE_AT_COLL_YR").TX_WEATHER_NL.value_counts(normalize=True).unstack()
a.plot.bar()
```

```python
# geen relatie tussen wegomstandigheid en de leeftijd van de wagen
a = collision_df.groupby("TX_CLASS_VEHICLE_AGE_AT_COLL_YR").TX_ROAD_CONDITION_NL.value_counts(normalize=True).unstack()
a.plot.bar()
```

```python
# nat en sneeuw maken het gevaarlijker bij schemering

a = collision_df.groupby("TX_ROAD_CONDITION_NL").TX_LIGHT_CONDITION_NL.value_counts(normalize=True).unstack()
a.plot.bar()
```

```python
a = collision_df.groupby("CD_LIGHT_CONDITION").TX_VEHICLE_SEGMENT_B_NL.value_counts(normalize=True).unstack()
a.plot()
```

```python
# bij weinig verlichting, de ongevallen worden zwaarder
a = collision_df.groupby("TX_LIGHT_CONDITION_NL").TX_ACCIDENT_CLASS_NL.value_counts(normalize=True).unstack()
a.plot(rot=-90)
```

```python
# overdag en in de nacht met verlichting meeste ongevallen in absolute aantallen 
# -> in deze situaties rijden de meeste wagens
# -> interessant om te normaliseren
a = collision_df.groupby("TX_LIGHT_CONDITION_NL").TX_ACCIDENT_CLASS_NL.value_counts().unstack()
a.plot(rot=-90)
```

```python
# afwezigheid van openbare verlichting komt meer voor buiten bebouwde kom
a = collision_df.groupby("TX_LIGHT_CONDITION_NL").TX_BUILD_UP_AREA_NL.value_counts(normalize=True).unstack()
a.plot(rot=-90) 
```

```python
# Namen en luxemburg meer ongevallen snechts met geen verlichting
a = collision_df[collision_df["CD_ROAD_TYPE"] == 1].groupby("TX_PROV_COLLISION_NL").TX_LIGHT_CONDITION_NL.value_counts(normalize=True).unstack()
a.plot.bar(rot=-90)
```

```python
a = collision_df.groupby("TX_REGION_COLLISON_NL").TX_LIGHT_CONDITION_NL.value_counts(normalize=True).unstack()
a.plot.bar(rot=-90)
```

```python
# moto's hebben opvallend gedrag
# - vaker ongeval met maar één bestuurder betrokken en geen hindernissen (gewooon ter val gekomen)
# - veel minder vaak botsing langs achteren, eerder in de zijkant inrijden
# - vaker ongeval met hindernis op baan
# - veel minder ongevallen met voetganger
# - veel minder kettingbotsingen
#   -> kan beter ontwijken
# met sportswagens en stadswagens vaker tegen hindernissen buiten de baan
a = collision_df.groupby("TX_VEHICLE_SEGMENT_B_NL").TX_COLLISON_TYPE_NL.value_counts(normalize=True).unstack()
a
b = a.plot.bar(rot=-90) 
# b.set_xticklabels(collision_df['TX_VEHICLE_SEGMENT_B_NL'].unique()[:7])
collision_df['TX_VEHICLE_SEGMENT_B_NL'].unique()
```

```python
# autosnelweg
# - vaker bosting langs achteren
# - veel minder frontaal (logisch)
# - veel minder langs opzij (logisch)
# - veel minder met voetganger (logisch)
# - veel meer kettingbotsingen (ook veel wagens achter elkaar)
# - minder vaak met maar één bestuurder
# gemeenteweg
# - meer met voetgangers
# - minder kettingbotsingen
# - minder botsingen langs achteren
a = collision_df.groupby("TX_ROAD_TYPE_NL").TX_COLLISON_TYPE_NL.value_counts(normalize=True).unstack()
a.plot.bar(rot=-90) 
```

```python
# iets meer ongevallen 's nachts op de snelweg dan op andere wegen
a = collision_df.groupby("TX_ROAD_TYPE_NL").TX_LIGHT_CONDITION_NL.value_counts(normalize=True).unstack()
a.plot.bar(rot=-90) 
```

```python
a = collision_df.groupby("TX_LIGHT_CONDITION_NL").TX_ROAD_TYPE_NL.value_counts(normalize=True).unstack()
a.plot.bar(rot=-90) 
```

```python
# zonder openbare verlichting meer botsingen met hindernissen buiten rijbaan
# - geen probleem op plaatsen met verlichting maar niet ontstoken (platsen waar uit onderzoek is gebleken dat verlichting niet zo nodig is)
# - meer ongevallen zonder hindernissen
# ook meer botsingen met hindernissen op de rijbaan als geen verlichting
# als verlichting niet ontstoken, meer botsingen langs achteren
# hoe meer licht, hoe meer de botsingen langs de zijkant plaatsvinden
# zones zonder verlichting minder ongevallen met voetgangers, zijn ook monder bewandelbare zones
a = collision_df.groupby("TX_LIGHT_CONDITION_NL").TX_COLLISON_TYPE_NL.value_counts(normalize=True).unstack()
a.plot.bar(rot=-90) 
```
