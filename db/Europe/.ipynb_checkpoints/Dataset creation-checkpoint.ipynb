# Infrastructure investments

Dit deel gaat over het geld gespendeerd aan wegen 

```python
#COUNTRY CODE CONVERTER

# Define a function to convert country codes to country names
def get_country_name(code):
    try:
        return pycountry.countries.get(alpha_3=code).name
    except:
        return None
```

```python
#dataset infrastructure investments
Investdataset = pd.read_csv(InvestmentFileName)
InvestdataTemp = Investdataset[Investdataset["SUBJECT"] == "ROAD"][Investdataset["TIME"]== 2019]
# Convert country codes to country


# Add a new column to the DataFrame with the country names
InvestdataTemp['Country'] = InvestdataTemp['LOCATION'].apply(lambda x: get_country_name(x))


InvestdataTemp = InvestdataTemp[InvestdataTemp['Country'].isin(european_countries)].reset_index(drop=True)

#Get the GDP dataset
GDPdataset = pd.read_csv(GDPFileName)
GDPdatasetTemp = GDPdataset[GDPdataset['Country'].isin(european_countries)]
GDPdatasetTemp["log(GDP(2019))"] = np.log(GDPdatasetTemp['GDP(2019)'])


#Now creating the GDP per investment dataset
InvestPerGDPTemp = pd.merge(InvestdataTemp, GDPdatasetTemp, on='Country', how='outer')

InvestPerGDPTemp["log(GDP(2019))"] = np.log(InvestPerGDPTemp['GDP(2019)'])
InvestPerGDPTemp
#InvestPerGDPVal = InvestPerGDPTemp["Value"]/ InvestPerGDPTemp["GDP(2019)"]

#InvestPerGDPTemp["InvestPerGDPVal"] = InvestPerGDPVal



Investdata = InvestPerGDPTemp[['Country', "Value"]].reset_index(drop=True)#InvestPerGDPTemp[['Country', "Value", "InvestPerGDPVal"]].reset_index(drop=True)
Investdata = Investdata.rename(columns={'Value': 'Invest'})


#Investdata
#InvestdataTemp
```

# Inwoners en oppervlakte

```python
areaDataset = pd.read_csv(areaFileName)
areaDataset.rename(columns={"country_name" : "Country"}, inplace=True)
areaDataset = areaDataset[["Country","area_km2"]]

#areaDataset

def code_to_country(code):
    try:
        country = pycountry.countries.get(alpha_2=code)
        return country.name
    except:
        return "Invalid country code"

popDataset = pd.read_csv(popFileName)

# Create a list of valid country codes
valid_country_codes = [country.alpha_2 for country in pycountry.countries]

# Convert country codes to country names
popDataset = popDataset[popDataset['geo'].isin(valid_country_codes)]

popDataset['Country'] = popDataset['geo'].apply(code_to_country)

popDataset = popDataset[popDataset["TIME_PERIOD"] == 2019]
popDataset.rename(columns={"OBS_VALUE" : "pop"}, inplace=True)


popDataset = popDataset[["Country","pop"]]
popDataset
```
